```python
import pandas as pd

# Load the dataset
df = pd.read_csv('ncr_ride_bookings.csv')
```

```python
# Print the general information about the dataframe, to evaluate if it is suitable for the assignment
print("Dataframe")
df.info()
```

```python
# Display the first 5 rows of the dataset
print("First 5 Rows")
print(df.head().to_markdown(numalign="left", stralign="left"))
```

```python
# Display the last 5 rows of the dataset
print("Last 5 Rows")
print(df.tail().to_markdown(numalign="left", stralign="left"))
```

```python
# Select the columns that we're going to keep for our analysis
columns_to_keep = [
    'Date',
    'Time',
    'Booking ID',
    'Booking Status',
    'Customer ID',
    'Vehicle Type',
    'Pickup Location',
    'Drop Location',
    'Booking Value',
    'Ride Distance',
    'Driver Ratings',
    'Customer Rating',
    'Payment Method'
]
```

```python
# Create a new dataframe with only the selected columns
df_filtered = df[columns_to_keep]
```

```python
# Filter out out rows where the 'Booking Status' is "No Driver Found"
df_filtered = df_filtered[df_filtered['Booking Status'] != 'No Driver Found']
```

```python
# Drop rows where key metrics columns (Booking value and Ride distance) have missing values
df_cleaned = df_filtered.dropna(subset=['Booking Value', 'Ride Distance'])
```

```python
# Display the information of the new, cleaned DataFrame
print("Cleaned Dataframe Info")
df_cleaned.info()
```

```python
# Generate summary statistics (for numerical columns)
print("--- Summary Statistics for Numerical Columns ---")
print(df_cleaned.describe().to_markdown(numalign="left", stralign="left"))
```

```python
# Identify Data Quality Issues
print("Null values ")
print(df_cleaned.isnull().sum().to_markdown(numalign="left", stralign="left"))

print("Data types")
print(df_cleaned.info())
```

```python
# Check if there are any duplicate rows
num_duplicates = df_cleaned.duplicated().sum()
print(f"Duplicate rows: {num_duplicates}")
```

```python
# Check if there are any relevant outliers in numerical columns
print("Outliers")
print(df_cleaned[['Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating']].describe().to_markdown(numalign="left", stralign="left"))
```

```python
# Correlation table for the numerical columns
print("Correlation matrix for numerical columns")

numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df_cleaned[numerical_cols].corr()
print(correlation_matrix.to_markdown(numalign="left", stralign="left"))
```
